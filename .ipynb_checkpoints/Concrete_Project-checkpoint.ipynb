```python
# Predição da Resistência à Compressão do Concreto

Neste projeto, vamos analisar o dataset "Concrete Compressive Strength" do UCI Machine Learning Repository para entender os fatores que influenciam a resistência à compressão do concreto e construir um modelo de regressão linear múltipla para prever essa resistência.
```

```python
## Objetivos do Projeto

- Realizar a importação e limpeza dos dados.
- Explorar os dados através de visualizações e estatísticas descritivas.
- Analisar as correlações entre variáveis.
- Construir e validar um modelo de regressão linear múltipla.
- Interpretar os resultados e extrair insights para a engenharia civil.
```

```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

# Configurando o estilo dos gráficos para melhor visualização
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Carregando o dataset real do concreto
# Certifique-se de que o arquivo 'Concrete_Data.xls' esteja na mesma pasta do Notebook
df = pd.read_excel("Concrete_Data.xls")

# Visualizando as primeiras linhas do dataset
df.head()
```

```python
# Verificando valores nulos
print("Valores nulos por coluna:")
print(df.isnull().sum())

# Estatísticas descritivas
print("\nEstatísticas Descritivas:")
print(df.describe())

# Dimensões do dataset
print("\nDimensões do dataset:", df.shape)
```

```python
# Plotando histogramas para visualizar a distribuição das variáveis
df.hist(bins=20, color='skyblue', edgecolor='black', figsize=(14,10))
plt.suptitle("Distribuição das Variáveis", fontsize=16)
plt.show()
```

```python
# Calculando a matriz de correlação
correlation_matrix = df.corr()

# Criando o heatmap com a paleta de cores "BuGn"
sns.heatmap(correlation_matrix, annot=True, cmap="BuGn", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlação")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
# Scatter plot: Resistência vs. Cimento

sns.scatterplot(x="Cement (component 1)(kg in a m^3 mixture)", y="Concrete compressive strength(MPa, megapascals) ", data=df)
plt.title("Resistência vs. Cimento")
plt.xlabel("Cimento (kg/m³)")
plt.ylabel("Resistência (MPa)")
plt.show()

# Scatter plot: Resistência vs. Água
sns.scatterplot(x="Water  (component 4)(kg in a m^3 mixture)", y="Concrete compressive strength(MPa, megapascals) ", data=df)
plt.title("Resistência vs. Água")
plt.xlabel("Água (kg/m³)")
plt.ylabel("Resistência (MPa)")
plt.show()
```

```python
# Teste de normalidade (Shapiro-Wilk) para a variável alvo
stat, p = stats.shapiro(df['Concrete compressive strength(MPa, megapascals) '])
print("Teste de Normalidade para 'Concrete compressive strength': estatística=%.3f, p=%.3f" % (stat, p))
```

```python
# Calculando a correlação de Pearson entre Cimento e Resistência
corr_cement_strength, _ = stats.pearsonr(df['Cement (component 1)(kg in a m^3 mixture)'], df['Concrete compressive strength(MPa, megapascals) '])
print("Correlação entre Cimento e Resistência:", round(corr_cement_strength, 2))
```

```python
# Selecionando as variáveis independentes


X = df[["Cement (component 1)(kg in a m^3 mixture)", "Blast Furnace Slag (component 2)(kg in a m^3 mixture)", "Fly Ash (component 3)(kg in a m^3 mixture)", "Water  (component 4)(kg in a m^3 mixture)", "Superplasticizer (component 5)(kg in a m^3 mixture)",
        "Coarse Aggregate  (component 6)(kg in a m^3 mixture)", "Fine Aggregate (component 7)(kg in a m^3 mixture)", "Age (day)"]]
X = sm.add_constant(X)  # Adiciona o intercepto ao modelo
y = df["Concrete compressive strength(MPa, megapascals) "]

# Ajustando o modelo de regressão linear múltipla
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Calculando os resíduos do modelo
residuals = model.resid

# Histograma dos resíduos
sns.histplot(residuals, kde=True, color="purple")
plt.title("Distribuição dos Resíduos")
plt.xlabel("Resíduos")
plt.ylabel("Frequência")
plt.show()

# Scatter plot dos resíduos vs. valores ajustados
fitted_vals = model.fittedvalues
sns.scatterplot(x=fitted_vals, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Valores Ajustados")
plt.ylabel("Resíduos")
plt.title("Resíduos vs. Valores Ajustados")
plt.show()
```

## Conclusões e Insights

- **Correlação:** As variáveis como "Cement" mostram forte relação com a resistência à compressão.
- **Modelo de Regressão:** O modelo ajustado apresenta um R² satisfatório, e os coeficientes com p-valor < 0.05 indicam variáveis estatisticamente significativas.
- **Validação:** A análise dos resíduos sugere que os pressupostos da regressão linear foram atendidos.

Esses resultados podem auxiliar engenheiros civis na otimização de projetos, identificando quais insumos têm maior impacto na resistência do concreto.
